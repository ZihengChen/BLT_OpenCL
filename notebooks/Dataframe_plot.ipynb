```python
import pandas as pd
from pylab import *
import glob

# df = pd.read_hdf("../data/hdf5/EGamma_2018B/15FAE8CF-39D9-2743-BFB2-F271F517D89D.h5")
files = glob.glob('../data/hdf5/EGamma_2018B/*.h5')
df = pd.concat([ pd.read_hdf(file) for file in files],ignore_index=True)
```

```python
# plot
configs = [
    ( "Dilepton_mass", 40,140, 2, r"$M_{e_1 e_2}$ [GeV]"),
    ( "Leptons_deltaR", 0,7, 0.2, r"$\Delta R_{e_1 e_2}$")
    ]

for config in configs:
    v,a,b,step,xl = config
    mybin = np.arange(a,b,step)
    
    plt.figure(figsize=(6,4),facecolor='w')
    plt.hist(df[v], mybin, color='grey', label="Run2018B (Partial)")
    plt.xlabel(xl)
    plt.legend(loc="upper right")
    plt.legend(loc="upper right")
    plt.grid(linestyle="--",color='grey',alpha=0.4)
    plt.savefig("../plots/{}.png".format(v),dpi=200)
```
